```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
data = pd.read_csv("NewspaperData.csv")
print(data)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
#Boxplot for daily column

plt.figure(figsize=(6,3))
plt.title("Box plot for Daily Sales")
plt.boxplot(data["daily"], vert = False)
plt.show()
```

```python
sns.histplot(data['daily'], kde = True, stat='density')
plt.show()
```

```python
#Boxplot for sunday column
plt.figure(figsize=(6,3))
plt.title("Box plot for Sunday Sales")
plt.boxplot(data["sunday"], vert = False)
plt.show()
```

```python
sns.histplot(data['sunday'], kde = True, stat='density')
plt.show()
```

## obsevations

- There are no missing values
- The daily column values appears to be right-skewed
- The sunday column values also to be right-skwed
- There are two outliers in both daily column and sunday column as observed for boxplots

## scatter Plot and Correlation Strength

```python
x = data["daily"]
y = data["sunday"]
plt.scatter(data["daily"], data["sunday"])
plt.xline(0, max(x) + 100)
plt.yline(0, max(x) + 100)
plt.show()
```

```python
data["daily"].corr(data["sunday"])
```

```python
data[["daily","sunday"]].corr()
```

## Obsevations
- The relationship between x(daily) and y(sunday) is seen to be linear as seen from scatter plot
- The correlations is strong positive with Pearson's correlation of 0.958154

## Fit a Linear Regression Model

```python
import statsmodels.formula.api as smf
model1 = smf.ols("sunday-daily",data = data).fit()
```
